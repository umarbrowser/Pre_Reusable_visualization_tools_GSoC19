```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline

y = np.random.randn(1000)
plt.plot(y)
```

```python
x = np.linspace(-10., 10., 1000)
y = np.sin(3*x) * np.exp(-.1 * x**2)
plt.plot(x,y)
```

```python
x = np.linspace(-5., 5., 100)
y = np.sin(3 * x) * np.exp(-.1 * x**2)
plt.plot(x,y, '--^', lw=3, color='#f95e4f', mfc='#5c898c', ms=8)
```

```python
x = np.random.randn(100)
y = x + np.random.randn(100)
plt.scatter(x, y)
```

```python
# Left
plt.subplot(1,2,1)
x = np.linspace(-10., 10., 1000)
plt.plot(x,np.sin(x), '-r', label='sinus')
plt.plot(x,np.cos(x), ':g', lw=1, label='cosinus')
plt.xticks([-10,0,10])
plt.yticks([1,0,10])
plt.ylim(-2,2)
plt.ylabel('x axis')
plt.xlabel('x axis')
plt.title('Two axis')
plt.legend()

# Right
plt.subplot(1,2,2, polar=True)
x = np.linspace(0, 2 * np.pi, 1000)
plt.plot(x, 1 + 2 * np.cos(6 * x))
plt.yticks([])
plt.xlim(-.1, 3.1)
plt.ylim(-.1, 3.1)
plt.xticks(np.linspace(0,5 * np.pi / 3, 6))
plt.title('A plot')
plt.grid(color='k', linewidth=1, linestyle=':')
```
